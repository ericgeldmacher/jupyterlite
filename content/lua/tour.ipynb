# Printing

the way the printing is done can be configured.
By default we use a pretty printer different from the default print

```lua
-- by default we use a nice pretty-printer
pprint("print pretty:")
a_table = {table_key=1}
print(a_table)

--to use luas default 
print("\nprint default:")
ilua.config.printer = "print"
print( a_table)

--to use pretty printer again
print("\nprint pretty again:")
ilua.config.printer = "pprint"
print(a_table)
```

## Last value printing
when executing a cell, xeus-lua tries to print the statement.

```lua

-- this will be printed since print({1,1,1}) is an expression
{1,1,1}
-- aaa
```

```lua
-- this will not be printend since a = {1,1,1} is not a valid expression
a = {1,1,1}
```

```lua
function fobar()
    return "this is fobar"
end
```

```lua
-- the function result will be printed
fobar()
```

this automatic printing behavior can be disabled

```lua
ilua.config.auto_print = false
```

```lua
-- the function result will **not** be printed
fobar()
```

```lua
-- set back to the defaults
ilua.config.auto_print = true
```

# transient data and metadata 
## display id

```lua
json = ilua.display.json({a=1, b="string", c={1,2,3}})
metadata = {} -- empty metadata
transient = {display_id = "id1"} 
ilua.display.display_data(json, metadata, transient)
```

```lua
json = ilua.display.json({a=1, b="updated string", c={1,2,3}})
metadata = {} -- empty metadata
transient = {display_id = "id1"} 
ilua.display.update_display_data(json, metadata, transient)
```
